```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_style('whitegrid')
```

```python
df = pd.read_csv('gtdb_soil_plasmids_taxon_intersection.tsv', sep='\t')

df.head()
```

```python
melted_df = pd.melt(df, id_vars='level')
```

```python
fig, ax = plt.subplots()
sns.barplot(data=melted_df, x='level', y='value', hue='variable', ax=ax)
ax.set_yscale('log')
for c in ax.containers:
    ax.bar_label(c, fmt=lambda x: int(x), label_type='edge')
```
