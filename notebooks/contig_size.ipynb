```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = {}
with open('IMG_assemblies/lengths.txt') as f:
    data['IMG'] = [int(i.strip()) for i in f]

with open('mgnify_assemblies/lengths.txt') as f:
    data['MGnify'] = [int(i.strip()) for i in f]
```

```python
sns.histplot(data, bins=50, binrange=[2_000,30_000])
```

```python
print(np.mean(data['IMG']), np.mean(data['MGnify']))
```

```python
print(np.median(data['IMG']), np.median(data['MGnify']))
```
